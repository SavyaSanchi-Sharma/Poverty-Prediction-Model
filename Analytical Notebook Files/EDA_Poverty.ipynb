```python
Poverty_1 = df[[ 'country', 'year','hc_ratio','international_poverty_gap', '$1_poverty_gap']].reset_index().drop(['index'],axis=1)
Poverty_1
```

```python
#sns.set(style="whitegrid")
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
sns.scatterplot(data=Poverty_1, x='hc_ratio', y='international_poverty_gap', ax=ax[0, 0])
sns.scatterplot(data=Poverty_1, x='hc_ratio', y='$1_poverty_gap', ax=ax[0, 1])
ax[0, 1].set_title('$1 Poverty Gap vs. HC Ratio')
ax[0, 0].set_title('International Poverty Gap vs. HC Ratio')
ax[1, 0].axis('off')
ax[1, 1].axis('off')
plt.tight_layout()
plt.show()
```

```python


def country_time_poverty_gap(country):
    country_data = df[df['country'] == country]
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=country_data, x='year', y='hc_ratio', marker='o')
    plt.title(f'HC Ratio Over Time for {country}')
    plt.xlabel('Year')
    plt.ylabel('HC Ratio')
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=country_data, x='year', y='international_poverty_gap', marker='o')
    plt.title(f'International Poverty Gap Over Time for {country}')
    plt.xlabel('Year')
    plt.ylabel('International Poverty Gap')
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=country_data, x='year', y='$1_poverty_gap', marker='o')
    plt.title(f'1$ Poverty gap Over Time for {country}')
    plt.xlabel('Year')
    plt.ylabel('1$ Poverty Gap')
    plt.grid()
    plt.show()
    
```

```python
country_time_poverty_gap('India')
```

```python
Poverty_2 = df[[ 'country', 'year','headcount_international_povline','total_shortfall_international_povline']].reset_index().drop(['index'],axis=1)
Poverty_2
```

```python
fig=plt.figure(figsize=(8,8))
sns.scatterplot(data=Poverty_2, x='headcount_international_povline', y='total_shortfall_international_povline')
plt.tight_layout()
plt.ylabel('Total Shortfall according to international poverty line')
plt.xlabel('Head Count ratio')
plt.show()
```

```python
def country_time_shortfall(country):
    country_data = df[df['country'] == country]
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=country_data, x='year', y='headcount_international_povline', marker='o')
    plt.title(f'HC  Over Time for {country}')
    plt.xlabel('Year')
    plt.ylabel('HC ')
    plt.grid()
    plt.show()
    
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=country_data, x='year', y='total_shortfall_international_povline', marker='o')
    plt.title(f'Total shortfall according to the international poverty line Gap Over Time for {country}')
    plt.xlabel('Year')
    plt.ylabel('Total Shortfall')
    plt.grid()
    plt.show()
```
